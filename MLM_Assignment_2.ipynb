**1.1) Write a Python Program to implement your own myreduce() function which works exactly like
Python's built-in function reduce**

```
def myreduce(funct, lst):
  x=lst[0]
  for i in lst[1:]:
    x=funct(x,i)
  return x


def mul(a,b):
  return a*b

l1=[1,2,3,4,5]
myreduce(mul, l1)
```

**Task1) 3) Implement a function longestWord() that takes a list of words and returns the longest one.**


```
l1=["anuj", "nitant", "sanky","sairaj", "prathamesh"]
def longestWord(lst):
  large=lst[0]
  for i in range(1, len(lst)):
    if(len(lst[i])>len(large)):
      large=lst[i]
  print(large)
longestWord(l1)
```

**Task 2) 1.1)
Write a Python Program(with class concepts) to find the area of the triangle using the below
formula.
area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
Function to take the length of the sides of triangle from user should be defined in the parent
class and function to calculate the area should be defined in subclass.**

```
class Side:
  def __init__(self, a, b, c):
    self.a=a
    self.b=b
    self.c=c

  def userInput(self):
    self.a=int(input("Enter left side:"))
    self.b=int(input("Enter down side:"))
    self.c=int(input("Enter right side:"))

class Area(Side):
  def __init__(self, *args):
    super(Side, self).__init__(*args)

  def calculate(self):
    s=(self.a+self.b+self.c)/2
    return (s*(s-self.a)*(s-self.b)*(s-self.c)) ** 0.5

a = Area()
a.userInput()
a.calculate()

```

**Task 2) 1.2) Write a function filter_long_words() that takes a list of words and an integer n and returns the list
of words that are longer than n.**


```
l1=["anuj", "nitant", "sanky", "sairaj", "prathamesh"]
def filter_long_words(lst, n):
  for i in range(0, len(lst)):
    if(len(lst[i])>n):
      print(lst[i])


filter_long_words(l1, 5)
```
