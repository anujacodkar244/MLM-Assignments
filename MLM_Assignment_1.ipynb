1) Write a program which will find all such numbers which are divisible by 7 but are not a multiple
of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a
comma-separated sequence on a single line.

```
for i in range(2000,3201):
    if(i%7==0 and i%5!=0):
        print(i, end=",")
```

2) Write a Python program to accept the user's first and last name and then getting them printed in
the the reverse order with a space between first name and last name.

```
fn=input("Enter 1st name:")
ln=input("Enter last name:")
i=len(fn)-1
j=len(ln)-1
rev_fn=""
rev_ln=""
while i!=-1:
  rev_fn=rev_fn+fn[i]
  i=i-1

while j!=-1:
  rev_ln=rev_ln+ln[j]
  j=j-1

print("Reversed Name: "+rev_fn+" "+rev_ln)
```

2) Create the below pattern using nested for loop in Python.
#*
# * *
#* * *
#* * * *
#* * * * *
#* * * *
#* * *
#* *
#* 

```
for i in range(0, 4, 1):
  for j in range(0, i+1):
    print("*",end="")
  print("\n")

for i in range(4, -1, -1):
  for j in range(0, i+1):
    print("*",end="")
  print("\n")
```

3) Write a Python program to reverse a word after accepting the input from the user.


```
pk=input("Enter word:")
l=len(pk)-1
rev=""
while l!=-1:
  rev=rev+pk[l]
  l=l-1

print(rev)
```
